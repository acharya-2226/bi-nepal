# 🎤 Project Presentation Template Mapping

This section maps each presentation point to the current implementation files using paths relative to this notebook location (`documentations/`).

---

## 🎯 1. Problem Definition and Objective

Backend objective and API description are centered in [main.py](../ecommerce_backend/main.py), where app setup and startup bootstrap are defined.  

The high-level project scope is documented in [PROJECT_DOCUMENTATION.md](../ecommerce_backend/PROJECT_DOCUMENTATION.md).  

Analytics scope is exposed through [analysis.py](../ecommerce_backend/routes/analysis.py), covering profitability, discounts, inventory, products, expenses, growth, breakeven, and cashflow endpoints.

---

## 📂 2. Dataset Description and Data Source

Excel is the primary ingestion source handled in [stage_01_ingest.py](../ecommerce_backend/pipeline/stage_01_ingest.py), while dataset expectations are summarized in [PROJECT_DOCUMENTATION.md](../ecommerce_backend/PROJECT_DOCUMENTATION.md).  

Persisted storage is SQLite, implemented in [database.py](../ecommerce_backend/database.py) through `get_connection`, `init_db`, `merge_context`, and `load_store`.

---

## 📥 3. Data Collection and Ingestion Method

Upload and validation are implemented in [upload.py](../ecommerce_backend/routes/upload.py).  

Pipeline orchestration is handled by [runner.py](../ecommerce_backend/pipeline/runner.py).  

Startup auto-bootstrap flow is defined in [main.py](../ecommerce_backend/main.py) inside the startup event logic.

---

## 🧹 4. Data Cleaning and Feature Engineering

Cleaning rules (missing value handling, type normalization, row filtering) are implemented in [stage_02_clean.py](../ecommerce_backend/pipeline/stage_02_clean.py).  

Derived metrics and shared variables are built in [stage_03_transform.py](../ecommerce_backend/pipeline/stage_03_transform.py).
